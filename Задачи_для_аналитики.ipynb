```python
from scipy import stats

group_a = [85, 90, 78, 92]
group_b = [92, 95, 89, 96]

_, p_a = stats.normaltest(group_a)
_, p_b = stats.normaltest(group_b)

if p_a > 0.05 and p_b > 0.05:
  stat, p_value = stats.ttest_ind(p_a, p_b)
  print(p_value)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Загрузка данных
df = pd.read_csv('dataset.tsv', sep='\t', header=None,
                 names=['day', 'test_id', 'feature_index', 'feature_value'])

# Создаем словарь для хранения данных по фичам
control_data = {i: [] for i in range(1500)}
test_data = {i: [] for i in range(1500)}

# Собираем данные по фичам
for _, row in df.iterrows():
    feat_idx = row['feature_index']
    value = row['feature_value']

    if row['test_id'] == 1:
        control_data[feat_idx].append(value)
    else:
        test_data[feat_idx].append(value)

# Список для хранения результатов (индекс фичи, p-value)
results = []

# Обработка каждой фичи
for feat_idx in range(1500):
    c_data = control_data[feat_idx]
    t_data = test_data[feat_idx]

    # Пропускаем фичи с недостаточными данными
    if len(c_data) < 10 or len(t_data) < 10:
        results.append((feat_idx, 1.0))  # Незначимое p-value
        continue

    # Выполняем KS-тест
    try:
        _, p_value = stats.ks_2samp(c_data, t_data)
        results.append((feat_idx, p_value))
    except:
        results.append((feat_idx, 1.0))

# Сортируем по p-value (по возрастанию значимости)
results.sort(key=lambda x: x[1])

# Выбираем топ-10 наиболее значимых фичей
top_10 = sorted([feat_idx for feat_idx, p_value in results[:10]])

# Форматируем вывод
print(','.join(map(str, top_10)))
```
